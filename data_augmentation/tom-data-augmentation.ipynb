```python
import csv
import os
import pickle
import random

import numpy as np


def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo)
    return dict


X = []
Y = []
Z = []
ids = []
O = []
subclass_generalclass_map = {}
imageNetId_taskId_map = {}
taskId_className_map = {}
with open("data/subclass_mapping.csv", "r", newline="", encoding="utf-8") as f:
    csv_file = csv.reader(f)
    next(csv_file, None)

    for row in csv_file:
        try:
            ids.append(int(row[4]))
            subclass_generalclass_map[int(row[4])] = int(row[3])
            imageNetId_taskId_map[int(row[4])] = int(row[0])
            taskId_className_map[int(row[0])] = row[1]
        except ValueError:
            pass
    taskId_className_map[87] = "Novel"

X_dog = []
Y_dog = []

X_bird = []
Y_bird = []

X_reptile = []
Y_reptile = []

subclass_count_map = {}

all_x = []
all_y = []

counts = [0, 0, 0]  # [bird, dog, reptile]

max_size = 36224
def load_databatch(data_folder, idx, img_size=32):
    data_file = os.path.join(data_folder, 'train_data_batch_')

    d = unpickle(data_file + str(idx))
    x = d['data']
    y = d['labels']
    x = np.dstack((x[:, :1024], x[:, 1024:2048], x[:, 2048:]))
    x = x.reshape((x.shape[0], 32, 32, 3))
    all_x.extend(x)
    all_y.extend(y)
    for i_1 in range(len(y)):
        if y[i_1] in ids:
            # this is to make sure bird or dog only has max of 35931 images
            if subclass_generalclass_map[y[i_1]] == 0:
                if counts[0] >= max_size:
                    continue
                counts[0] += 1
            if subclass_generalclass_map[y[i_1]] == 1:
                if counts[1] >= max_size:
                    continue
                counts[1] += 1
            if subclass_generalclass_map[y[i_1]] == 2:
                if counts[2] >= max_size:
                    continue
                counts[2] += 1
            if y[i_1] not in subclass_count_map:
                subclass_count_map[y[i_1]] = 0
            subclass_count_map[y[i_1]] += 1
            X.append(x[i_1])
            O.append(y[i_1])
            Y.append(imageNetId_taskId_map[y[i_1]])
            Z.append(subclass_generalclass_map[y[i_1]])
            # if subclass_generalclass_map[y[i]] == 0:
            #     X_bird.append(x[i])
            #     Y_bird.append(y[i])
            # elif subclass_generalclass_map[y[i]] == 1:
            #     X_dog.append(x[i])
            #     Y_dog.append(y[i])
            # else:
            #     X_reptile.append(x[i])
            #     Y_reptile.append(y[i])


for i in range(1, 11):
    load_databatch("../imagenet", i)

# randomly add max_size novel images
indices = list(range(len(all_x)))
random.shuffle(indices)
counter = 0
for i in indices:
    if counter >= max_size:
        break
    if all_y[i] not in ids:
        X.append(all_x[i])
        Y.append(87)
        Z.append(3)
        O.append(-1)
        counter += 1

# shuffle X, Y, Z
indices = list(range(len(X)))
random.shuffle(indices)
X_save = []
Y_save = []
Z_save = []
O_save = []
for i in indices:
    X_save.append(X[i])
    Y_save.append(Y[i])
    Z_save.append(Z[i])
    O_save.append(O[i])

with open('data' + '.pkl', 'wb') as file:
    # print('bird', len(X_bird)) # 37343
    # print('dog', len(X_dog)) # 37377
    # print('reptile', len(X_reptile)) # 35931
    print(subclass_count_map)
    pickle.dump(
        {'Images': X_save, 'Labels': Y_save, 'GeneralClass': Z_save},
        file)
```
